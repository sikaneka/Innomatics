Running Sum of 1d Array

```python
class Solution(object):
    def runningSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        output=[]
        output.append(nums[0])
        for i in range(len(nums)-1):
            output.append(output[i]+nums[i+1])
        return output
```

shuffle the array

```python
class Solution(object):
    def shuffle(self, nums, n):
        """
        :type nums: List[int]
        :type n: int
        :rtype: List[int]
        """
        mid=(len(nums)/2)-1
        l1=nums[:mid+1]
        l2=nums[mid+1:]
        result=[]
        for i in range(len(l1)):
            result.append(l1[i])
            result.append(l2[i])
        return result
        
```

Kids with Greatest Number of candies

```python
class Solution(object):
    def kidsWithCandies(self, candies, extraCandies):
        """
        :type candies: List[int]
        :type extraCandies: int
        :rtype: List[bool]
        """
        result=[]
        max_candy=max(candies)
        for i in candies:
            if i+extraCandies>=max_candy:
                result.append(True)
            else:
                result.append(False)
        return result
        
```
