Number of Steps to Reduce a Number to Zero

```python
class Solution(object):
    def numberOfSteps(self, num):
        """
        :type num: int
        :rtype: int
        """
        result=0
        while num>0:
            if num%2==0:
                result+=1
                num/=2
            else:
                result+=1
                num-=1
        return result


        
```
