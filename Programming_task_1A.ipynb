Loops

```python
import sys
if __name__ == '__main__':
    n = int(sys.stdin.readline().strip())
    if 1<=n<=20:
        for i in range(n):
            print(i*i)
    else:
        print("the number doessn't included in the constraint")
```

Write a function

```python
def is_leap(year):
    leap = False
    
    if ((year%4==0) and (year%100!=0)) or (year%400==0):
        leap=True
    
    return leap

year = int(input())
print(is_leap(year))
```

```python
import sys
if __name__ == '__main__':
    n = int(sys.stdin.readline().strip())
    if 1<=n<=150:
        num=""
        for i in range(1,n+1):
            num=num+str(i)
        print(num)
```
